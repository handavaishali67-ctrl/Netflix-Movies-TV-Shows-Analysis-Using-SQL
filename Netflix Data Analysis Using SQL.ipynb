## **Netflix Data Analysis Using SQL**

**Dataset Link**: https://www.kaggle.com/datasets/shivamb/netflix-shows?resource=download

#### Set Catalog & Schema

```sql
USE CATALOG netflix_project;
USE SCHEMA default;
```

#### Dataset Overview and Sample Records

```sql
SELECT *
FROM netflix_titles
LIMIT 10;
```

#### Total Number of Records

```sql
SELECT COUNT(*) 
FROM netflix_titles
```

#### List All Column Names

```sql
DESCRIBE netflix_titles;
```

#### Count NULLs in Each Column

```sql
SELECT
  COUNT(*) AS total_rows,
  SUM(CASE WHEN show_id IS NULL THEN 1 ELSE 0 END) AS show_id_nulls,
  SUM(CASE WHEN type IS NULL THEN 1 ELSE 0 END) AS type_nulls,
  SUM(CASE WHEN title IS NULL THEN 1 ELSE 0 END) AS title_nulls,
  SUM(CASE WHEN director IS NULL THEN 1 ELSE 0 END) AS director_nulls,
  SUM(CASE WHEN cast IS NULL THEN 1 ELSE 0 END) AS cast_nulls,
  SUM(CASE WHEN country IS NULL THEN 1 ELSE 0 END) AS country_nulls,
  SUM(CASE WHEN date_added IS NULL THEN 1 ELSE 0 END) AS date_added_nulls,
  SUM(CASE WHEN release_year IS NULL THEN 1 ELSE 0 END) AS release_year_nulls,
  SUM(CASE WHEN rating IS NULL THEN 1 ELSE 0 END) AS rating_nulls,
  SUM(CASE WHEN duration IS NULL THEN 1 ELSE 0 END) AS duration_nulls,
  SUM(CASE WHEN listed_in IS NULL THEN 1 ELSE 0 END) AS listed_in_nulls,
  SUM(CASE WHEN description IS NULL THEN 1 ELSE 0 END) AS description_nulls
FROM netflix_titles;
```

```sql
SELECT
  COUNT(*) AS total_rows,
  ROUND(100 * SUM(CASE WHEN director IS NULL THEN 1 ELSE 0 END) / COUNT(*), 2) AS director_null_pct,
  ROUND(100 * SUM(CASE WHEN country IS NULL THEN 1 ELSE 0 END) / COUNT(*), 2) AS country_null_pct
FROM netflix_titles;
```

#### Create a Clean Table/View

```sql
CREATE OR REPLACE VIEW netflix_clean AS
SELECT
  show_id,
  type,
  title,
  IFNULL(director, 'Unknown') AS director,
  IFNULL(cast, 'Unknown') AS cast,
  IFNULL(country, 'Unknown') AS country,
  date_added,
  release_year,
  IFNULL(rating, 'Not Rated') AS rating,
  duration,
  listed_in,
  description
FROM netflix_titles
WHERE title IS NOT NULL
  AND type IS NOT NULL;
```

#### Re-check NULLs

```sql
SELECT
  COUNT(*) AS total_rows,
  COUNT(director) AS director_not_null,
  COUNT(country) AS country_not_null
FROM netflix_clean;
```

#### 1) Movies vs TV Shows

```sql
SELECT type, COUNT(*) AS total_titles
FROM netflix_clean
GROUP BY type;
```

#### 2) Titles Released Per Year

```sql
SELECT release_year, COUNT(*) AS total_titles
FROM netflix_clean
GROUP BY release_year
ORDER BY release_year DESC;
```

#### 3) Top 10 Countries

```sql
SELECT country, COUNT(*) AS total_titles
FROM netflix_clean
GROUP BY country
ORDER BY total_titles DESC
LIMIT 10;
```

#### 4) Rating Distribution

```sql
SELECT rating, COUNT(*) AS total_titles
FROM netflix_clean
GROUP BY rating
ORDER BY total_titles DESC;
```

#### 5) Most Common Genres

```sql
SELECT listed_in, COUNT(*) AS total_titles
FROM netflix_clean
GROUP BY listed_in
ORDER BY total_titles DESC
LIMIT 10;
```

#### 6) Simple Duration Insight

```sql
SELECT type, COUNT(*) AS total
FROM netflix_clean
WHERE duration LIKE '%min%'
GROUP BY type;
```

#### 7) Movies vs TV Shows by Year

```sql
SELECT
  release_year,
  type,
  COUNT(*) AS total_titles
FROM netflix_clean
GROUP BY release_year, type
ORDER BY release_year DESC;
```

#### 8) Top Directors

```sql
SELECT director, COUNT(*) AS total_titles
FROM netflix_clean
WHERE director <> 'Unknown'
GROUP BY director
ORDER BY total_titles DESC
LIMIT 10;
```

#### 9) Content Added After 2018

```sql
SELECT COUNT(*) AS total_titles
FROM netflix_clean
WHERE release_year >= 2019;
```

#### 10) Long Movies vs Short Movies

```sql
SELECT
  COUNT(*) AS long_movies
FROM netflix_clean
WHERE type = 'Movie'
  AND duration LIKE '%min%';
```

#### 11) TV Shows by Seasons

```sql
SELECT duration, COUNT(*) AS total_shows
FROM netflix_clean
WHERE type = 'TV Show'
GROUP BY duration
ORDER BY total_shows DESC;
```

#### 12) Titles with Missing Country Info

```sql
SELECT COUNT(*) AS unknown_country_titles
FROM netflix_clean
WHERE country = 'Unknown';
```

#### 13) Titles by Rating & Type

```sql
SELECT rating, type, COUNT(*) AS total_titles
FROM netflix_clean
GROUP BY rating, type
ORDER BY total_titles DESC;
```

#### 14) Search-Style Query

```sql
SELECT title, type, release_year
FROM netflix_clean
WHERE title LIKE '%Love%'
ORDER BY release_year DESC;
```
